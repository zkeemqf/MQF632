```python
# ============================================
# CELL 1: Import Libraries and Setup
# ============================================

import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Machine Learning Libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import optuna

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("‚úÖ Libraries imported successfully")

# ============================================
# CELL 2: Load All Three Momentum Data Sources
# ============================================

def load_and_preprocess_data():
    """Load and preprocess all three momentum data sources"""
    
    print("Loading momentum data sources...")
    
    # 1. Load price momentum data
    price_momentum = pd.read_csv('hmm_enhanced_momentum_scores_2020_2024.csv')
    price_momentum['date'] = pd.to_datetime(price_momentum['date'])
    print(f"‚úì Price momentum loaded: {price_momentum.shape}")
    
    # 2. Load news sentiment data
    news_sentiment = pd.read_csv('combined_sentiment_analysis_reduced.csv')
    news_sentiment['Date'] = pd.to_datetime(news_sentiment['Date'], format='%Y%m%d')
    news_sentiment.rename(columns={'Date': 'date', 'Ticker': 'ticker'}, inplace=True)
    print(f"‚úì News sentiment loaded: {news_sentiment.shape}")
    
    # 3. Load macro sentiment data
    macro_wide = pd.read_csv('Equity_Macro_Sentiments.csv')
    print(f"‚úì Macro sentiment loaded (wide format): {macro_wide.shape}")
    
    # Convert macro sentiment from wide to long format
    ticker_col = macro_wide.columns[0]
    macro_long = macro_wide.melt(
        id_vars=[ticker_col],
        var_name='date_str',
        value_name='macro_momentum'
    )
    
    # Rename ticker column if needed
    if ticker_col != 'ticker':
        macro_long.rename(columns={ticker_col: 'ticker'}, inplace=True)
    
    # Parse dates from format like 'May-25' to datetime
    def parse_date(date_str):
        try:
            parts = date_str.split('-')
            if len(parts) == 2:
                month_str, year_str = parts
                year = int('20' + year_str) if len(year_str) == 2 else int(year_str)
                month_map = {
                    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
                    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
                }
                month = month_map.get(month_str, 1)
                # Use last day of month
                day = 28 if month == 2 else (30 if month in [4, 6, 9, 11] else 31)
                return pd.Timestamp(year=year, month=month, day=day)
        except:
            return pd.NaT
    
    macro_long['date'] = macro_long['date_str'].apply(parse_date)
    macro_sentiment = macro_long.dropna(subset=['date', 'macro_momentum'])
    
    print(f"‚úì Macro sentiment converted to long format: {macro_sentiment.shape}")
    
    # Filter all data for 2022-2024 period
    start_date = '2022-01-01'
    end_date = '2024-12-31'
    
    price_momentum = price_momentum[(price_momentum['date'] >= start_date) & 
                                    (price_momentum['date'] <= end_date)]
    news_sentiment = news_sentiment[(news_sentiment['date'] >= start_date) & 
                                   (news_sentiment['date'] <= end_date)]
    macro_sentiment = macro_sentiment[(macro_sentiment['date'] >= start_date) & 
                                     (macro_sentiment['date'] <= end_date)]
    
    # Get common tickers across all three datasets
    common_tickers = set(price_momentum['ticker'].unique()) & \
                    set(news_sentiment['ticker'].unique()) & \
                    set(macro_sentiment['ticker'].unique())
    
    print(f"\nNumber of common tickers across all datasets: {len(common_tickers)}")
    print(f"Common tickers: {sorted(list(common_tickers))[:10]}...")  # Show first 10
    
    # Filter for common tickers
    price_momentum = price_momentum[price_momentum['ticker'].isin(common_tickers)]
    news_sentiment = news_sentiment[news_sentiment['ticker'].isin(common_tickers)]
    macro_sentiment = macro_sentiment[macro_sentiment['ticker'].isin(common_tickers)]
    
    return price_momentum, news_sentiment, macro_sentiment, sorted(list(common_tickers))

# Load all data
price_momentum, news_sentiment, macro_sentiment, tickers = load_and_preprocess_data()

# ============================================
# CELL 3: Three-Factor Feature Engineering
# ============================================

def create_momentum_features(price_df, sentiment_df, macro_df):
    """Create momentum features from price, sentiment, and macro data"""
    
    print("\nCreating three-factor features...")
    
    # 1. Aggregate sentiment scores by date and ticker
    sentiment_agg = sentiment_df.groupby(['date', 'ticker']).agg({
        'positive_prob': 'mean',
        'negative_prob': 'mean',
        'neutral_prob': 'mean',
        'sentiment_label': 'mean',
        'Relevance_Score': 'mean'
    }).reset_index()
    
    # Create sentiment momentum score
    sentiment_agg['sentiment_momentum'] = (
        sentiment_agg['positive_prob'] - sentiment_agg['negative_prob']
    ) * sentiment_agg['Relevance_Score']
    
    # 2. Merge price and sentiment
    merged_df = pd.merge(
        price_df[['date', 'ticker', 'momentum_score', 'state', 'eps_signal', 
                  'eps_change', 'price_eps_alignment']],
        sentiment_agg[['date', 'ticker', 'sentiment_momentum', 'positive_prob', 
                      'negative_prob', 'sentiment_label']],
        on=['date', 'ticker'],
        how='inner'
    )
    
    # 3. Convert macro sentiment to daily frequency (forward fill monthly to daily)
    print("Converting macro sentiment to daily frequency...")
    macro_daily = []
    
    for ticker in macro_df['ticker'].unique():
        ticker_macro = macro_df[macro_df['ticker'] == ticker].copy()
        ticker_macro = ticker_macro.set_index('date')
        
        if len(ticker_macro) > 0:
            # Create daily date range
            date_range = pd.date_range(start=ticker_macro.index.min(), 
                                      end=ticker_macro.index.max(), 
                                      freq='D')
            
            # Reindex to daily and forward fill
            ticker_macro = ticker_macro.reindex(date_range, method='ffill')
            ticker_macro['ticker'] = ticker
            ticker_macro = ticker_macro.reset_index()
            ticker_macro.rename(columns={'index': 'date'}, inplace=True)
            
            macro_daily.append(ticker_macro)
    
    macro_daily_df = pd.concat(macro_daily, ignore_index=True)
    
    # 4. Merge with macro sentiment
    merged_df = pd.merge(
        merged_df,
        macro_daily_df[['date', 'ticker', 'macro_momentum']],
        on=['date', 'ticker'],
        how='inner'
    )
    
    print(f"After merging three factors: {merged_df.shape}")
    
    # Sort by ticker and date
    merged_df = merged_df.sort_values(['ticker', 'date']).reset_index(drop=True)
    
    # 5. Initialize technical feature columns
    feature_columns = [
        'price_momentum_ma5', 'price_momentum_ma20',
        'sentiment_momentum_ma5', 'sentiment_momentum_ma20',
        'macro_momentum_ma5', 'macro_momentum_ma20',
        'price_momentum_change', 'sentiment_momentum_change', 'macro_momentum_change',
        'price_momentum_std', 'sentiment_momentum_std', 'macro_momentum_std'
    ]
    
    for col in feature_columns:
        merged_df[col] = 0.0
    
    # 6. Create technical features for each ticker
    for ticker in tickers:
        ticker_mask = merged_df['ticker'] == ticker
        ticker_indices = merged_df[ticker_mask].index
        
        if len(ticker_indices) > 0:
            # Moving averages
            merged_df.loc[ticker_indices, 'price_momentum_ma5'] = \
                merged_df.loc[ticker_indices, 'momentum_score'].rolling(5, min_periods=1).mean().values
            merged_df.loc[ticker_indices, 'price_momentum_ma20'] = \
                merged_df.loc[ticker_indices, 'momentum_score'].rolling(20, min_periods=1).mean().values
            
            merged_df.loc[ticker_indices, 'sentiment_momentum_ma5'] = \
                merged_df.loc[ticker_indices, 'sentiment_momentum'].rolling(5, min_periods=1).mean().values
            merged_df.loc[ticker_indices, 'sentiment_momentum_ma20'] = \
                merged_df.loc[ticker_indices, 'sentiment_momentum'].rolling(20, min_periods=1).mean().values
            
            merged_df.loc[ticker_indices, 'macro_momentum_ma5'] = \
                merged_df.loc[ticker_indices, 'macro_momentum'].rolling(5, min_periods=1).mean().values
            merged_df.loc[ticker_indices, 'macro_momentum_ma20'] = \
                merged_df.loc[ticker_indices, 'macro_momentum'].rolling(20, min_periods=1).mean().values
            
            # Momentum changes (using diff to avoid infinity)
            merged_df.loc[ticker_indices, 'price_momentum_change'] = \
                merged_df.loc[ticker_indices, 'momentum_score'].diff().values
            merged_df.loc[ticker_indices, 'sentiment_momentum_change'] = \
                merged_df.loc[ticker_indices, 'sentiment_momentum'].diff().values
            merged_df.loc[ticker_indices, 'macro_momentum_change'] = \
                merged_df.loc[ticker_indices, 'macro_momentum'].diff().values
            
            # Volatility
            merged_df.loc[ticker_indices, 'price_momentum_std'] = \
                merged_df.loc[ticker_indices, 'momentum_score'].rolling(20, min_periods=1).std().values
            merged_df.loc[ticker_indices, 'sentiment_momentum_std'] = \
                merged_df.loc[ticker_indices, 'sentiment_momentum'].rolling(20, min_periods=1).std().values
            merged_df.loc[ticker_indices, 'macro_momentum_std'] = \
                merged_df.loc[ticker_indices, 'macro_momentum'].rolling(20, min_periods=1).std().values
    
    # 7. Clean data
    merged_df = merged_df.fillna(0)
    merged_df = merged_df.replace([np.inf, -np.inf], 0)
    
    # Clip extreme values
    numeric_cols = merged_df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        if col not in ['date', 'state', 'eps_signal']:
            upper_bound = merged_df[col].quantile(0.999)
            lower_bound = merged_df[col].quantile(0.001)
            merged_df[col] = merged_df[col].clip(lower=lower_bound, upper=upper_bound)
    
    return merged_df

# Create features
features_df = create_momentum_features(price_momentum, news_sentiment, macro_sentiment)
print(f"\n‚úÖ Three-factor features created. Shape: {features_df.shape}")
print("Available columns:", features_df.columns.tolist())

# ============================================
# CELL 4: Target Variable and LSTM Sequences
# ============================================

def create_target_variable(df, forward_days=5):
    """Create target variable (future returns)"""
    
    # Try to load actual price data
    try:
        price_data = pd.read_csv('OHLCV_20201224_to_20241224_stooq.csv')
        price_data['Date'] = pd.to_datetime(price_data['Date'])
        price_data = price_data.rename(columns={'Date': 'date', 'Ticker': 'ticker'})
        
        # Merge with features
        df = pd.merge(df, price_data[['date', 'ticker', 'Close']], 
                     on=['date', 'ticker'], how='left')
        
        # Calculate forward returns for each ticker
        for ticker in tickers:
            ticker_mask = df['ticker'] == ticker
            ticker_data = df[ticker_mask].copy()
            
            # Forward returns
            ticker_data['target'] = ticker_data['Close'].shift(-forward_days) / ticker_data['Close'] - 1
            
            df.loc[ticker_mask, 'target'] = ticker_data['target']
        
        # Drop rows with NaN targets
        df = df.dropna(subset=['target'])
        print("‚úì Using actual price data for targets")
        
    except:
        print("‚ö†Ô∏è Using synthetic target based on three momentum factors")
        # Create synthetic target with all three momentum factors
        df['target'] = (
            0.5 * df['momentum_score'] + 
            0.3 * df['sentiment_momentum'] + 
            0.2 * df['macro_momentum'] + 
            np.random.normal(0, 0.01, len(df))
        )
    
    return df

def create_lstm_features(df, sequence_length=20):
    """Create LSTM sequences from time series data"""
    
    # Include all three momentum factors in features
    feature_cols = [
        'momentum_score', 'sentiment_momentum', 'macro_momentum',
        'price_momentum_ma5', 'sentiment_momentum_ma5', 'macro_momentum_ma5',
        'price_momentum_change', 'sentiment_momentum_change', 'macro_momentum_change',
        'eps_signal', 'price_eps_alignment'
    ]
    
    # Check available features
    available_features = [col for col in feature_cols if col in df.columns]
    print(f"Using {len(available_features)} features for LSTM")
    
    sequences = []
    targets = []
    static_features = []
    dates = []
    tickers_list = []
    
    for ticker in tickers:
        ticker_data = df[df['ticker'] == ticker].copy()
        
        if len(ticker_data) < sequence_length:
            continue
        
        # Clean and normalize features
        ticker_features = ticker_data[available_features].values
        ticker_features = np.nan_to_num(ticker_features, nan=0.0, posinf=0.0, neginf=0.0)
        
        scaler = StandardScaler()
        try:
            scaled_features = scaler.fit_transform(ticker_features)
        except:
            continue
        
        # Create sequences
        for i in range(sequence_length, len(ticker_data)):
            sequences.append(scaled_features[i-sequence_length:i])
            targets.append(ticker_data.iloc[i]['target'])
            
            # Static features now include all three momentum volatilities
            static_feat_cols = [
                'state', 'positive_prob', 'negative_prob',
                'price_momentum_std', 'sentiment_momentum_std', 'macro_momentum_std'
            ]
            available_static = [col for col in static_feat_cols if col in ticker_data.columns]
            
            static_feat = ticker_data.iloc[i][available_static].values
            static_features.append(static_feat)
            
            dates.append(ticker_data.iloc[i]['date'])
            tickers_list.append(ticker)
    
    return (np.array(sequences), np.array(targets), np.array(static_features), 
            dates, tickers_list)

# Create target and sequences
features_df = create_target_variable(features_df)
sequences, targets, static_features, dates, ticker_labels = create_lstm_features(features_df)

print(f"\n‚úÖ LSTM sequences created:")
print(f"   Sequences shape: {sequences.shape}")
print(f"   Static features shape: {static_features.shape}")
print(f"   Number of samples: {len(sequences)}")

# ============================================
# CELL 5: Hybrid LSTM + Random Forest Model
# ============================================

class HybridLSTMRandomForest:
    def __init__(self, sequence_length, n_lstm_features, n_static_features, 
                 lstm_units=64, rf_params=None):
        self.sequence_length = sequence_length
        self.n_lstm_features = n_lstm_features
        self.n_static_features = n_static_features
        self.lstm_units = lstm_units
        self.rf_params = rf_params or {}
        
        # Build LSTM feature extractor
        self.lstm_model = self._build_lstm_extractor()
        
        # Random Forest will be created during fit
        self.rf_model = None
        self.scaler = StandardScaler()
        
    def _build_lstm_extractor(self):
        """Build LSTM model for feature extraction"""
        model = Sequential([
            LSTM(self.lstm_units, return_sequences=True, 
                 input_shape=(self.sequence_length, self.n_lstm_features)),
            Dropout(0.2),
            LSTM(self.lstm_units // 2, return_sequences=False),
            Dropout(0.2),
            Dense(32, activation='relu'),
            Dense(16, activation='relu')
        ])
        
        model.compile(optimizer=Adam(0.001), loss='mse')
        return model
    
    def extract_lstm_features(self, sequences):
        """Extract features using LSTM"""
        lstm_features = Model(inputs=self.lstm_model.input, 
                            outputs=self.lstm_model.layers[-1].output)
        return lstm_features.predict(sequences, batch_size=32)
    
    def fit(self, X_sequences, X_static, y, momentum_weights=None):
        """Fit the hybrid model"""
        
        # Train LSTM to predict targets
        self.lstm_model.fit(X_sequences, y, epochs=20, batch_size=32, 
                          validation_split=0.2, verbose=0)
        
        # Extract LSTM features
        lstm_features = self.extract_lstm_features(X_sequences)
        
        # Apply momentum weights if provided (for 3 factors)
        if momentum_weights is not None and len(momentum_weights) == 3:
            X_static_weighted = X_static.copy()
            if X_static.shape[1] >= 3:
                X_static_weighted[:, 0] *= momentum_weights[0]  # Price
                X_static_weighted[:, 1] *= momentum_weights[1]  # Sentiment
                X_static_weighted[:, 2] *= momentum_weights[2]  # Macro
            X_static = X_static_weighted
        
        # Combine LSTM features with static features
        combined_features = np.hstack([lstm_features, X_static])
        
        # Scale features
        combined_features_scaled = self.scaler.fit_transform(combined_features)
        
        # Train Random Forest
        self.rf_model = RandomForestRegressor(**self.rf_params)
        self.rf_model.fit(combined_features_scaled, y)
        
        return self
    
    def predict(self, X_sequences, X_static, momentum_weights=None):
        """Make predictions"""
        
        # Extract LSTM features
        lstm_features = self.extract_lstm_features(X_sequences)
        
        # Apply momentum weights if provided
        if momentum_weights is not None and len(momentum_weights) == 3:
            X_static_weighted = X_static.copy()
            if X_static.shape[1] >= 3:
                X_static_weighted[:, 0] *= momentum_weights[0]
                X_static_weighted[:, 1] *= momentum_weights[1]
                X_static_weighted[:, 2] *= momentum_weights[2]
            X_static = X_static_weighted
        
        # Combine features
        combined_features = np.hstack([lstm_features, X_static])
        
        # Scale features
        combined_features_scaled = self.scaler.transform(combined_features)
        
        # Predict
        return self.rf_model.predict(combined_features_scaled)

print("‚úÖ Model class defined")

# ============================================
# CELL 6: Hyperparameter Optimization for Three Factors
# ============================================

def objective(trial):
    """Optuna objective function for three-factor optimization"""
    
    # Three momentum weights that sum to 1
    price_weight = trial.suggest_float('price_momentum_weight', 0.1, 0.7)
    sentiment_weight = trial.suggest_float('sentiment_momentum_weight', 0.1, 0.7)
    
    # Ensure weights are valid
    if price_weight + sentiment_weight >= 0.95:
        return -np.inf
    
    macro_weight = 1.0 - price_weight - sentiment_weight
    
    if macro_weight < 0.05:
        return -np.inf
    
    rf_params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 300),
        'max_depth': trial.suggest_int('max_depth', 5, 30),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
        'random_state': 42
    }
    
    lstm_units = trial.suggest_categorical('lstm_units', [32, 64, 128])
    
    # Time series cross-validation
    tscv = TimeSeriesSplit(n_splits=3)
    scores = []
    
    for train_idx, val_idx in tscv.split(sequences):
        # Split data
        X_seq_train, X_seq_val = sequences[train_idx], sequences[val_idx]
        X_static_train, X_static_val = static_features[train_idx], static_features[val_idx]
        y_train, y_val = targets[train_idx], targets[val_idx]
        
        # Create and train model
        model = HybridLSTMRandomForest(
            sequence_length=20,
            n_lstm_features=sequences.shape[2],
            n_static_features=static_features.shape[1],
            lstm_units=lstm_units,
            rf_params=rf_params
        )
        
        momentum_weights = [price_weight, sentiment_weight, macro_weight]
        
        try:
            model.fit(X_seq_train, X_static_train, y_train, momentum_weights)
            predictions = model.predict(X_seq_val, X_static_val, momentum_weights)
            score = -mean_squared_error(y_val, predictions)  # Negative MSE
            scores.append(score)
        except:
            return -np.inf
    
    return np.mean(scores)

# Run optimization
print("üîÑ Starting three-factor hyperparameter optimization...")
print("This will take approximately 15-30 minutes...")

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)

# Extract best parameters
best_params = study.best_params
price_weight = best_params['price_momentum_weight']
sentiment_weight = best_params['sentiment_momentum_weight']
macro_weight = 1.0 - price_weight - sentiment_weight

print(f"\n‚úÖ Optimization complete!")
print(f"\nüéØ Optimal three-factor weights:")
print(f"   Price Momentum: {price_weight:.1%}")
print(f"   Sentiment Momentum: {sentiment_weight:.1%}")
print(f"   Macro Momentum: {macro_weight:.1%}")

# ============================================
# CELL 7: Train Final Model
# ============================================

# Best RF parameters
rf_params = {
    'n_estimators': best_params['n_estimators'],
    'max_depth': best_params['max_depth'],
    'min_samples_split': best_params['min_samples_split'],
    'min_samples_leaf': best_params['min_samples_leaf'],
    'max_features': best_params['max_features'],
    'random_state': 42
}

# Train-test split
train_size = int(0.8 * len(sequences))
X_seq_train, X_seq_test = sequences[:train_size], sequences[train_size:]
X_static_train, X_static_test = static_features[:train_size], static_features[train_size:]
y_train, y_test = targets[:train_size], targets[train_size:]
dates_train, dates_test = dates[:train_size], dates[train_size:]

# Train final model
print("\nüöÄ Training final three-factor model...")
final_model = HybridLSTMRandomForest(
    sequence_length=20,
    n_lstm_features=sequences.shape[2],
    n_static_features=static_features.shape[1],
    lstm_units=best_params['lstm_units'],
    rf_params=rf_params
)

momentum_weights = [price_weight, sentiment_weight, macro_weight]
final_model.fit(X_seq_train, X_static_train, y_train, momentum_weights)

# Make predictions
train_predictions = final_model.predict(X_seq_train, X_static_train, momentum_weights)
test_predictions = final_model.predict(X_seq_test, X_static_test, momentum_weights)

# Calculate performance
train_r2 = r2_score(y_train, train_predictions)
test_r2 = r2_score(y_test, test_predictions)
train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

print(f"\n‚úÖ Model Performance:")
print(f"   Train R¬≤: {train_r2:.4f}")
print(f"   Test R¬≤: {test_r2:.4f}")
print(f"   MSE Train: {train_mse:.6f}")
print(f"   MSE Test: {test_mse:.6f}")

# ============================================
# CELL 8: Evaluation and Visualization
# ============================================

# Feature importance
feature_names = [f'lstm_feat_{i}' for i in range(16)] + \
                ['state', 'positive_prob', 'negative_prob', 
                 'price_momentum_std', 'sentiment_momentum_std', 'macro_momentum_std']

feature_importance = pd.DataFrame({
    'feature': feature_names,
    'importance': final_model.rf_model.feature_importances_
}).sort_values('importance', ascending=False)

# Create visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Feature Importance
axes[0, 0].barh(feature_importance['feature'][:10], 
                feature_importance['importance'][:10])
axes[0, 0].set_xlabel('Importance')
axes[0, 0].set_title('Top 10 Feature Importances')

# 2. Predictions vs Actual
axes[0, 1].scatter(y_test, test_predictions, alpha=0.5)
axes[0, 1].plot([y_test.min(), y_test.max()], 
                [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0, 1].set_xlabel('Actual Returns')
axes[0, 1].set_ylabel('Predicted Returns')
axes[0, 1].set_title('Predictions vs Actual (Test Set)')

# 3. Residuals
residuals = y_test - test_predictions
axes[1, 0].hist(residuals, bins=50, edgecolor='black')
axes[1, 0].set_xlabel('Residuals')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('Residual Distribution')

# 4. Time series of predictions
test_df = pd.DataFrame({
    'date': dates_test,
    'actual': y_test,
    'predicted': test_predictions
})
test_df = test_df.sort_values('date')

axes[1, 1].plot(test_df['date'][:100], test_df['actual'][:100], 
                label='Actual', alpha=0.7)
axes[1, 1].plot(test_df['date'][:100], test_df['predicted'][:100], 
                label='Predicted', alpha=0.7)
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Returns')
axes[1, 1].set_title('Predictions Over Time (First 100 samples)')
axes[1, 1].legend()
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# ============================================
# CELL 9: Generate Alpha Signals
# ============================================

def generate_alpha_signals(model, sequences, static_features, momentum_weights, 
                          dates, tickers, features_df, threshold=0.01):
    """Generate comprehensive trading signals"""
    
    predictions = model.predict(sequences, static_features, momentum_weights)
    
    # Generate signals
    signals = np.zeros_like(predictions)
    signals[predictions > threshold] = 1  # Long
    signals[predictions < -threshold] = -1  # Short
    
    # Get momentum scores
    momentum_scores = []
    for i, (date, ticker) in enumerate(zip(dates, tickers)):
        row = features_df[(features_df['date'] == date) & (features_df['ticker'] == ticker)]
        if len(row) > 0:
            momentum_scores.append({
                'price_momentum': row.iloc[0]['momentum_score'],
                'sentiment_momentum': row.iloc[0]['sentiment_momentum'],
                'macro_momentum': row.iloc[0]['macro_momentum']
            })
        else:
            momentum_scores.append({
                'price_momentum': 0,
                'sentiment_momentum': 0,
                'macro_momentum': 0
            })
    
    # Create output dataframe
    output_df = pd.DataFrame({
        'date': dates,
        'ticker': tickers,
        'predicted_return': predictions,
        'signal': signals,
        'signal_strength': np.abs(predictions),
        'price_momentum_score': [m['price_momentum'] for m in momentum_scores],
        'sentiment_momentum_score': [m['sentiment_momentum'] for m in momentum_scores],
        'macro_momentum_score': [m['macro_momentum'] for m in momentum_scores],
        'price_momentum_weight': price_weight,
        'sentiment_momentum_weight': sentiment_weight,
        'macro_momentum_weight': macro_weight,
        'combined_momentum': sum([
            price_weight * np.array([m['price_momentum'] for m in momentum_scores]),
            sentiment_weight * np.array([m['sentiment_momentum'] for m in momentum_scores]),
            macro_weight * np.array([m['macro_momentum'] for m in momentum_scores])
        ]),
        'model_version': '3_factor_lstm_rf_final'
    })
    
    # Add rankings and z-scores
    output_df['daily_rank'] = output_df.groupby('date')['predicted_return'].rank(pct=True)
    output_df['prediction_zscore'] = output_df.groupby('date')['predicted_return'].transform(
        lambda x: (x - x.mean()) / x.std() if x.std() > 0 else 0
    )
    
    # Add quintile assignment
    output_df['quintile'] = output_df.groupby('date')['predicted_return'].transform(
        lambda x: pd.qcut(x, q=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'], duplicates='drop')
    )
    
    return output_df

# Generate signals for all data
print("\nüìä Generating alpha signals...")
all_predictions = final_model.predict(sequences, static_features, momentum_weights)
alpha_signals_df = generate_alpha_signals(
    final_model, sequences, static_features, momentum_weights,
    dates, ticker_labels, features_df, threshold=0.01
)

# Display summary
print("\nSignal Distribution:")
print(alpha_signals_df['signal'].value_counts())
print(f"\nTotal signals generated: {len(alpha_signals_df):,}")

# Show top signals
print("\nTop 10 Long Signals:")
top_long = alpha_signals_df[alpha_signals_df['signal'] == 1].nlargest(10, 'predicted_return')
print(top_long[['date', 'ticker', 'predicted_return', 'combined_momentum']])

print("\nTop 10 Short Signals:")
top_short = alpha_signals_df[alpha_signals_df['signal'] == -1].nsmallest(10, 'predicted_return')
print(top_short[['date', 'ticker', 'predicted_return', 'combined_momentum']])

# ============================================
# CELL 10: Export Results
# ============================================

# Generate timestamp
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# Export alpha signals
csv_filename = f'alpha_signals_3factor_{timestamp}.csv'
alpha_signals_df.to_csv(csv_filename, index=False)
print(f"\n‚úÖ Alpha signals exported to: {csv_filename}")

# Also save as parquet for efficiency
parquet_filename = csv_filename.replace('.csv', '.parquet')
alpha_signals_df.to_parquet(parquet_filename, index=False)
print(f"‚úÖ Also saved as parquet: {parquet_filename}")

# Create metadata
metadata = {
    'model_info': {
        'type': 'Three-Factor LSTM-RF Hybrid',
        'version': '1.0',
        'training_date': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    },
    'momentum_weights': {
        'price': float(price_weight),
        'sentiment': float(sentiment_weight),
        'macro': float(macro_weight)
    },
    'model_parameters': {
        'lstm_units': best_params['lstm_units'],
        'sequence_length': 20,
        'rf_n_estimators': rf_params['n_estimators'],
        'rf_max_depth': rf_params['max_depth'],
        'rf_min_samples_split': rf_params['min_samples_split'],
        'rf_min_samples_leaf': rf_params['min_samples_leaf'],
        'rf_max_features': rf_params['max_features']
    },
    'performance_metrics': {
        'train_r2': float(train_r2),
        'test_r2': float(test_r2),
        'train_mse': float(train_mse),
        'test_mse': float(test_mse)
    },
    'data_info': {
        'n_tickers': len(tickers),
        'n_samples': len(alpha_signals_df),
        'date_range': {
            'start': alpha_signals_df['date'].min().strftime('%Y-%m-%d'),
            'end': alpha_signals_df['date'].max().strftime('%Y-%m-%d')
        },
        'signal_distribution': {
            'long': int((alpha_signals_df['signal'] == 1).sum()),
            'short': int((alpha_signals_df['signal'] == -1).sum()),
            'neutral': int((alpha_signals_df['signal'] == 0).sum())
        }
    },
    'feature_importance': {
        'top_10_features': feature_importance.head(10).to_dict('records')
    }
}

# Save metadata
import json
metadata_filename = f'model_metadata_3factor_{timestamp}.json'
with open(metadata_filename, 'w') as f:
    json.dump(metadata, f, indent=4)
print(f"‚úÖ Metadata saved to: {metadata_filename}")

# ============================================
# CELL 11: Performance Summary
# ============================================

# Calculate additional metrics
ic_corr = np.corrcoef(test_predictions, y_test)[0, 1]
hit_rate = (np.sign(test_predictions) == np.sign(y_test)).mean()

# Create summary report
print("\n" + "="*70)
print("THREE-FACTOR ALPHA SIGNAL GENERATION COMPLETE")
print("="*70)

report = f"""
üìä Model Performance Summary:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚Ä¢ Train R¬≤: {train_r2:.4f}
‚Ä¢ Test R¬≤: {test_r2:.4f}
‚Ä¢ Information Coefficient: {ic_corr:.4f}
‚Ä¢ Hit Rate: {hit_rate:.2%}

‚öñÔ∏è Optimal Momentum Weights:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚Ä¢ Price Momentum: {price_weight:.1%}
‚Ä¢ Sentiment Momentum: {sentiment_weight:.1%}
‚Ä¢ Macro Momentum: {macro_weight:.1%}

üìà Signal Statistics:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚Ä¢ Total Signals: {len(alpha_signals_df):,}
‚Ä¢ Long Signals: {(alpha_signals_df['signal'] == 1).sum():,} ({(alpha_signals_df['signal'] == 1).mean():.1%})
‚Ä¢ Short Signals: {(alpha_signals_df['signal'] == -1).sum():,} ({(alpha_signals_df['signal'] == -1).mean():.1%})
‚Ä¢ Neutral: {(alpha_signals_df['signal'] == 0).sum():,} ({(alpha_signals_df['signal'] == 0).mean():.1%})

üìÅ Output Files:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚Ä¢ Alpha Signals: {csv_filename}
‚Ä¢ Parquet Format: {parquet_filename}
‚Ä¢ Model Metadata: {metadata_filename}

üéØ Key Insights:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
‚Ä¢ The model successfully integrated all three momentum factors
‚Ä¢ {tickers[0]} to {tickers[-1]} and {len(tickers)-2} other stocks analyzed
‚Ä¢ Date coverage: {alpha_signals_df['date'].min().strftime('%Y-%m-%d')} to {alpha_signals_df['date'].max().strftime('%Y-%m-%d')}

üöÄ Next Steps:
‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
1. Use the alpha signals CSV/Parquet file for GNN portfolio optimization
2. The 'combined_momentum' column provides the weighted momentum score
3. Use 'quintile' column for portfolio construction strategies
4. Monitor out-of-sample performance and retrain periodically
"""

print(report)

# Save report
report_filename = f'performance_report_3factor_{timestamp}.txt'
# open with utf-8 encoding so emojis and all Unicode work
with open(report_filename, 'w', encoding='utf-8') as f:
    f.write(report)
print(f"\n‚úÖ Performance report saved to: {report_filename}")


# ============================================
# CELL 12: Final Visualizations
# ============================================

# Create comprehensive final visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 1. Momentum Weights Pie Chart
axes[0, 0].pie([price_weight, sentiment_weight, macro_weight], 
               labels=['Price', 'Sentiment', 'Macro'],
               autopct='%1.1f%%', colors=['#FF6B6B', '#4ECDC4', '#45B7D1'])
axes[0, 0].set_title('Optimal Momentum Weights', fontsize=14, fontweight='bold')

# 2. Signal Distribution
signal_counts = alpha_signals_df['signal'].value_counts()
axes[0, 1].bar(['Short (-1)', 'Neutral (0)', 'Long (+1)'], 
               [signal_counts.get(-1, 0), signal_counts.get(0, 0), signal_counts.get(1, 0)],
               color=['red', 'gray', 'green'])
axes[0, 1].set_title('Signal Distribution', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('Count')

# 3. Prediction Distribution
axes[0, 2].hist(alpha_signals_df['predicted_return'], bins=50, edgecolor='black', alpha=0.7)
axes[0, 2].axvline(x=0.01, color='g', linestyle='--', label='Long threshold')
axes[0, 2].axvline(x=-0.01, color='r', linestyle='--', label='Short threshold')
axes[0, 2].set_title('Predicted Return Distribution', fontsize=14, fontweight='bold')
axes[0, 2].set_xlabel('Predicted Return')
axes[0, 2].legend()

# 4. Average Momentum by Signal Type
momentum_by_signal = alpha_signals_df.groupby('signal')[
    ['price_momentum_score', 'sentiment_momentum_score', 'macro_momentum_score']
].mean()

x = np.arange(len(momentum_by_signal))
width = 0.25

axes[1, 0].bar(x - width, momentum_by_signal['price_momentum_score'], 
               width, label='Price', color='#FF6B6B')
axes[1, 0].bar(x, momentum_by_signal['sentiment_momentum_score'], 
               width, label='Sentiment', color='#4ECDC4')
axes[1, 0].bar(x + width, momentum_by_signal['macro_momentum_score'], 
               width, label='Macro', color='#45B7D1')

axes[1, 0].set_xlabel('Signal Type')
axes[1, 0].set_title('Average Momentum Scores by Signal', fontsize=14, fontweight='bold')
axes[1, 0].set_xticks(x)
axes[1, 0].set_xticklabels(['Short', 'Neutral', 'Long'])
axes[1, 0].legend()

# 5. Top 10 Stocks by Signal Count
top_tickers = alpha_signals_df[alpha_signals_df['signal'] != 0]['ticker'].value_counts().head(10)
axes[1, 1].barh(top_tickers.index, top_tickers.values, color='steelblue')
axes[1, 1].set_title('Top 10 Active Stocks', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Number of Signals')

# 6. Combined Momentum Distribution
axes[1, 2].hist(alpha_signals_df['combined_momentum'], bins=50, 
                edgecolor='black', alpha=0.7, color='purple')
axes[1, 2].set_title('Combined Momentum Distribution', fontsize=14, fontweight='bold')
axes[1, 2].set_xlabel('Combined Momentum Score')
axes[1, 2].set_ylabel('Frequency')

plt.tight_layout()
plt.savefig(f'model_summary_3factor_{timestamp}.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚ú® All done! Your three-factor alpha signals are ready for GNN portfolio optimization.")
print("üìå Remember: The output file contains all momentum scores, weights, and signals needed for the next stage.")
```
